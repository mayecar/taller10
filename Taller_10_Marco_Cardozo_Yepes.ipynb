```python
#Vamos a contar las diferentes banderas (estaciones de servicio)
def contar(n, vec):
  '''Para esta funcion n es el numero a contar y vec es la lista donde contar '''
  count=0
  for i in vec:
    if i==n:
      count+=1
  return count
```

```python
print('TERPEL aparece: ',contar('TERPEL', lista_bandera))
print('SAVE aparece: ',contar('SAVE', lista_bandera))
print('BIOMAX aparece: ',contar('BIOMAX', lista_bandera))
print('TEXACO aparece: ',contar('TEXACO', lista_bandera))
print('AYATAWACOOP aparece: ',contar('AYATAWACOOP', lista_bandera))
print('PETROMIL aparece: ',contar('PETROMIL', lista_bandera))
print('ESSO aparece: ',contar('ESSO', lista_bandera))
print('COOMULPINORT aparece: ',contar('COOMULPINORT', lista_bandera))
print('MOBIL aparece: ',contar('MOBIL', lista_bandera))
print('ZEUSS aparece: ',contar('ZEUSS', lista_bandera))
print('BRIO aparece: ',contar('BRIO', lista_bandera))
print('ECOS aparece: ',contar('ECOS', lista_bandera))
print('DISCOWACOOP aparece: ',contar('DISCOWACOOP', lista_bandera))
print('ZAPATA Y VELASQUEZ aparece: ',contar('ZAPATA Y VELASQUEZ', lista_bandera))
print('PUMA aparece: ',contar('PUMA', lista_bandera))
print('GULF aparece: ',contar('GULF', lista_bandera))
print('CASAMOTOR aparece: ',contar('CASAMOTOR', lista_bandera))
print('PETROBRAS aparece: ',contar('PETROBRAS', lista_bandera))
print('OCTANO aparece: ',contar('OCTANO', lista_bandera))


aux=[contar('TERPEL', lista_bandera), contar('SAVE', lista_bandera),contar('BIOMAX', lista_bandera),
     contar('TEXACO', lista_bandera),contar('AYATAWACOOP', lista_bandera), contar('PETROMIL', lista_bandera),
     contar('ESSO', lista_bandera) , contar('COOMULPINORT', lista_bandera),contar('MOBIL', lista_bandera),
     contar('ZEUSS', lista_bandera),contar('BRIO', lista_bandera),contar('ECOS', lista_bandera),contar('DISCOWACOOP', lista_bandera),
     contar('ZAPATA Y VELASQUEZ', lista_bandera),contar('PUMA', lista_bandera),contar('GULF', lista_bandera),contar('CASAMOTOR', lista_bandera),
     contar('PETROBRAS', lista_bandera),contar('OCTANO', lista_bandera)]

print(aux)
print('--------------------------------')
print('La suma total es de: ',sum(aux))
```
